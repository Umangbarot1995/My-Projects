Q1 what is visualization. what is the purpose?  

Visualization is the representation of data and information in a graphical or pictorial form, such as charts, graphs, maps, and images. The purpose of visualization is to make complex data and information easier to understand and interpret. By using visual aids, patterns, trends, and relationships in data can be quickly identified and analyzed, and insights can be gained that might not be immediately apparent from raw data alone

Q2. How do you make effective visualization? 

Making effective visualizations involves several key steps:
    1. Define the purpose: Start by identifying the goal of the visualization and what insights you hope to communicate. This will help guide the design and choice of visualizations.
    2. Choose the right type of visualization: There are many types of visualizations, each with its own strengths and weaknesses. Consider the nature of the data and the message you want to communicate when choosing the appropriate visualization.
    3. Prepare the data: Ensure that the data is accurate, complete, and properly formatted. This may involve cleaning and transforming the data to make it more suitable for visualization.
    4. Design the visualization: Pay attention to the visual elements of the visualization, including the choice of colors, labels, axes, and scales. Consider the audience and the message you want to communicate when designing the visualization.
    5. Use appropriate scales: Ensure that the scales used in the visualization accurately represent the data. For example, using a logarithmic scale may be appropriate for data with a wide range of values, while a linear scale may be better for data with similar values.
    6. Label and annotate the visualization: Provide clear and concise labels and annotations to help the audience understand the data and the message being communicated.
    7. Test and refine the visualization: Once the visualization is complete, test it with a sample audience to ensure that it is effective in communicating the desired message. Refine the visualization as needed based on feedback.

Q12 Create the following using data in x and y 

```python
import numpy as np
import matplotlib.pyplot as plt

# create the figure and three subplots
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
x = np.linspace(0, 5, 11)


# plot something in the first subplot
axs[0].plot(x, x**2, lw=2, color="lightblue", label = "x**2")
axs[0].plot(x, x**3, lw=2, color="orange",  label = "x**3")
axs[0].set_xlim(0, 5.5)
axs[0].set_ylim(0, 120)
axs[0].legend()
axs[0].set_title('default axes ranges')

# plot something in the second subplot
axs[1].plot(x, x**2, lw=2, color="lightblue", label = "x**2")
axs[1].plot(x, x**3, lw=2, color="orange", label = "x**3")
axs[1].set_xlim(0, 5.5)
axs[1].set_ylim(0, 120)
axs[1].legend()
axs[1].set_title('tight axes')

# # plot something in the third subplot
axs[2].plot(x, x**1, lw=2, color="lightblue", label = "x**2")
axs[2].plot(x, x**3, lw=2, color="orange", label = "x**3")
axs[2].set_xlim(2, 5)
axs[2].set_ylim(0, 60)
axs[2].legend()
axs[2].set_title('custom axes range')
# # set the overall title for the figure
fig.suptitle('Multiple plots')

# show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data to plot
x = np.linspace(0, 5, 11)
y = np.sin(x)

# Create a new figure and axes for the main plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, x**2, lw=2, color="blue")
ax.set_xlabel('X_label_axes2')
ax.set_ylabel('Y_label_axes2')
ax.set_xlim(0, 5)
ax.set_ylim(0, 25)

# Create a new set of axes for the subplot inside the main plot
subax = ax.inset_axes([0.2, 0.4, 0.4, 0.3])
subax.plot(y, y**2, lw=2, color="red")
subax.set_xlabel('X_label_subaxes2')
subax.set_ylabel('Y_label_subaxes2')
subax.set_xlim(0, 20)
subax.set_ylim(0, 5)
subax.set_title('Axes 2 Title')


fig.suptitle('Axes 2 Title')

# Show the plot
plt.show()
```
