```python
#Q10 Answer - 1
import seaborn as sns
import matplotlib.pyplot as plt

df_titanic = sns.load_dataset('titanic')
male_data = df_titanic.loc[df_titanic['sex'] == 'male']
female_data = df_titanic.loc[df_titanic['sex'] == 'female']
sns.set_style('whitegrid')
sns.histplot(data=male_data, x='age', color='skyblue', alpha=0.9, multiple='layer')
sns.histplot(data=female_data, x='age', color='pink', alpha=0.9, multiple='layer')
plt.show()
```

```python
# Q10 Answer - 2
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Filter for passengers with known age and class
titanic = titanic.loc[titanic["age"].notna() & titanic["class"].notna()]

# Define custom color palette
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]

# Create a histogram with age distribution for all passengers, with different colors for each class
sns.histplot(data=titanic, x="age", hue="class", palette=colors, alpha=0.8, element="step")

# Set chart title and labels
plt.title("Age Distribution on Titanic by Passenger Class")
plt.xlabel("Age")
plt.ylabel("Frequency")

# Show the chart
plt.show()
```
