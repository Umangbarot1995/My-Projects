```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_titanic = pd.read_csv("titanic.csv")
```

```python
titanic = sns.load_dataset('titanic')
sns.boxplot(x='survived', y='age', data=titanic)
plt.show()
```

I have create a Box plot where we can see median age of survivours is slightly lower than than of non-survivours

```python
sns.countplot(x='class', hue='survived', data=titanic)
plt.show()
```

I have used Stack bar chart where we can see that the majority of non-survivors were in the third class, while the majority of survivors were in the first and second classes

```python
sns.countplot(x='sex', hue='survived', data=titanic)
plt.show()
```

Created Stack bar Chart to see that the majority of non-survivors were male, while the majority of survivors were female

```python
sns.scatterplot(x='fare', y='age', hue='survived', data=titanic)
plt.show()
```

Used Scatter Plot to figure out that there is no clear relationship between fare and age for survivors and non-survivors
